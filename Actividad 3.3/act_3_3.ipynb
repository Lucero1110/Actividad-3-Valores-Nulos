**GENERAR UNA FUNCIÓN QUE CARGUE TIPOS DE ARCHIVOS**

1. Primero se le pide al usuario cuál es el archivo que desea cargar(CSV, XLSX, JSON, HTML)

2. Se crea la función llamandola datos_cargar, se le ingresa el parametro y de ahí va comprando que tipo de archivo es, dependiendo que cómo finaliza el nombre del archivo.

3. Sí es cualquiera de esos archivo, lo convertirá en .csv y lo guardara con el nombre de conver a ___ to csv

4. devuelve df para leerlo

5. Se llama a la función 

```python
archivo = input("¿Cuál archivo desea cargar?")

def datos_cargar (archivo):
    if '.csv' in archivo:
        df = pd.read_csv(archivo)
        print(df.to_csv("conver_1_csv.csv"))
    elif '.xlsx' in archivo:
        df =  pd.read_excel(archivo)   
        print(df.to_csv("conver_2_excel.csv")) 
    elif '.json' in archivo:
        df = pd.read_json(archivo)
        print(df.to_csv("conver_3_json.csv")) 
    elif '.html' in archivo:
        df = pd.read_html(archivo)
        df = pd.concat(df, ignore_index=True)
        df
        print(df.to_csv("conver_4_html.csv"))

        
        

    return(df)

```
